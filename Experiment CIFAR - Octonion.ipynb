```julia
using MLDatasets
using ImageCore
using LinearAlgebra
using PyPlot
using Random
using Statistics
using Printf
using Quaternions
using JLD2, FileIO
using ProgressMeter
include("HyperECNNs.jl")
rng = MersenneTwister(1234);
```

```julia
function cifar2gray(x)
    return round.(Int,255*(0.2989*x[:,:,1,1] + 0.5870*x[:,:,2,1]+ 0.1140*x[:,:,3,1]))
end

function imnoise(img,noise_std = 0.1)
    return clamp.(img + noise_std*randn(rng, Float64, size(img)),0,1)
end

function gray2bin(x,Nbits=8)
    y = Array{Int64}(undef,length(x),Nbits)
    for i=1:length(x)
        y[i,:] = 2*digits(x[i], base=2, pad = Nbits).-1
    end
    return y, size(x)
end

function bin2gray(y, xsize)
    return reshape([Int(sum(((y[i,:].+1)/2).*2 .^(0:size(y,2)-1))) for i=1:size(y,1)],xsize)
end
```

```julia
# load full training set
train_x, train_y = CIFAR10.traindata();
```

```julia
x = cifar2gray(train_x[:,:,:,3]);
```

```julia
xc, xsize = gray2bin(x,8);
```

```julia
y = bin2gray(xc,xsize);
```

```julia
imshow(transpose(y),cmap=get_cmap("gray"), vmin=0, vmax=255)
norm(x-y)
```

```julia
P = 200
# Indexes of the fundamental memories
indFM = randperm(rng, size(train_x)[4])[1:P];
```

```julia
# Bipolar fundamental memory set
N = 32*32
U = Array{Real}(undef,N,8,P)
for i=1:P
    uc, usize = gray2bin(cifar2gray(train_x[:,:,:,indFM[i]]),8)
    U[:,:,i] = uc
end
```

```julia
imgx = cifar2gray(train_x[:,:,:,indFM[1]])
imshow(transpose(imgx),cmap=get_cmap("gray"), vmin=0, vmax=255)
title("Undistorted image")
```

```julia
imgxr = cifar2gray(imnoise(train_x[:,:,:,indFM[1]],20/255))
x, xsize = gray2bin(imgxr,8)
imshow(transpose(imgxr),cmap=get_cmap("gray"), vmin=0, vmax=255)
title("Corrupted image")
```

```julia
alpha = 20/(8*N)
beta = -20
y, Energy = HyperECNNs.Sync(HyperECNNs.LambdaInner,[1,1,1,1,1,1,1,1],HyperECNNs.SplitSign,nothing,U,x,alpha,beta)
imgy = bin2gray(y,xsize)
imshow(transpose(imgy),cmap=get_cmap("gray"), vmin=0, vmax=255)
title("Recalled Image -- Synchronous Update")
println((norm(imgx-imgxr),norm(imgx-imgy)))
(100*sum(abs.(imgx-imgxr))/N,100*sum(abs.(imgx-imgy))/N)
```

```julia
N = 32*32

alpha = 20/(8*N)
beta = -20

P = 200

Nsim = 100
NoiseValues = range(0,stop=150,step=5)/255
Nnoise = length(NoiseValues)

SuccessOctonion_Sync = zeros(Int64,(Nnoise,))
SuccessOctonion_Seq = zeros(Int64,(Nnoise,))

@showprogress for s = 1:Nsim
    # Generate the fundamental memory set!
    indFM = randperm(rng, size(train_x)[4])[1:P]
    U = Array{Real}(undef,N,8,P)
    for i=1:P
        uc, usize = gray2bin(cifar2gray(train_x[:,:,:,indFM[i]]),8)
        U[:,:,i] = uc
    end
    for i = 1:Nnoise
        # Generate a corrupted pattern
        imgx = cifar2gray(train_x[:,:,:,indFM[1]])
        imgxr = cifar2gray(imnoise(train_x[:,:,:,indFM[1]],NoiseValues[i]))
        x, xsize = gray2bin(imgxr,8)
        
        # Recall using synchronous update
        y, Energy = HyperECNNs.Sync(HyperECNNs.LambdaInner,[1,1,1,1,1,1,1,1],HyperECNNs.SplitSign,nothing,
            U,x,alpha,beta)
        imgy = bin2gray(y,xsize)
        Error = norm(imgx-imgy)
        if Error<1.e-4
            SuccessOctonion_Sync[i] = SuccessOctonion_Sync[i] + 1
        end
        
        # Recall using asynchronous update
        y, Energy = HyperECNNs.Seq(HyperECNNs.LambdaInner,[1,1,1,1,1,1,1,1],HyperECNNs.SplitSign,nothing,
            U,x,alpha,beta)
        imgy = bin2gray(y,xsize)
        Error = norm(imgx-imgy)
        if Error<1.e-4
            SuccessOctonion_Seq[i] = SuccessOctonion_Seq[i] + 1
        end
    end
    @save "JLDs/CIFAROctonion.jld2" SuccessOctonion_Sync SuccessOctonion_Seq NoiseValues Nsim
end
```

```julia
@load "JLDs/CIFAROctonion.jld2"
plot(255*NoiseValues,SuccessOctonion_Seq/Nsim, marker="o",label = "Synchronous")
plot(255*NoiseValues,SuccessOctonion_Sync/Nsim, marker="s",label = "Asynchronous")
title("Octonion-valued ECNNs")
xlabel("Standard Deviation of Gaussian Noise")
ylabel("Success Rate")
legend()
grid()
```
